```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
measures = []
with open('result.txt', 'r') as f:
    for line in f.readlines():
        a, b = map(int, line.split())
        measures.append(a/b)
```

```python
# Plotting the data
plt.plot(range(len(measures[100:])), measures[100:], marker='o', markersize=1)
plt.axhline(y=0.012408, color='r', linestyle='--', label='actual value')
plt.xlabel('Index')
plt.ylabel('Measure Value')
plt.title('Measure Values vs Index')
plt.show()
```
